# U.S. Medical Insurance Costs

In this project, a **CSV** file with medical insurance costs will be investigated using Python fundamentals. The goal with this project will be to analyze various attributes within **insurance.csv** to learn more about the patient information in the file and gain insight into potential use cases for the dataset.

```python
# Import required module
import csv
```

```python
# ----------------------
# Step 1: Load and clean the data
# ----------------------
insurance_data = []

with open("insurance.csv", mode="r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        # Convert numeric values from strings
        row["age"] = int(row["age"])
        row["bmi"] = float(row["bmi"])
        row["children"] = int(row["children"])
        row["charges"] = float(row["charges"])
        insurance_data.append(row)
```

```python
# ----------------------
# Step 2: Helper function to calculate averages
# ----------------------
def calculate_average(dataset, key):
    """Return the average value of a specific column/key."""
    total = sum(person[key] for person in dataset)
    return round(total / len(dataset), 2)
```

```python
# ----------------------
# Step 3: Basic statistics
# ----------------------
avg_age = calculate_average(insurance_data, "age")
avg_bmi = calculate_average(insurance_data, "bmi")
avg_charges = calculate_average(insurance_data, "charges")

print("=== Basic Dataset Summary ===")
print(f"Average Age: {avg_age} years")
print(f"Average BMI: {avg_bmi}")
print(f"Average Insurance Charges: ${avg_charges}\n")
```

```python
# ----------------------
# Step 4: Compare smokers vs non-smokers
# ----------------------
def smoker_vs_non_smoker(dataset):
    smokers = [p["charges"] for p in dataset if p["smoker"] == "yes"]
    non_smokers = [p["charges"] for p in dataset if p["smoker"] == "no"]
    
    avg_smoker = round(sum(smokers) / len(smokers), 2)
    avg_non_smoker = round(sum(non_smokers) / len(non_smokers), 2)
    difference = round(avg_smoker - avg_non_smoker, 2)
    
    return avg_smoker, avg_non_smoker, difference

smoker_avg, non_smoker_avg, diff = smoker_vs_non_smoker(insurance_data)
print("=== Smokers vs Non-Smokers ===")
print(f"Average charge for smokers: ${smoker_avg}")
print(f"Average charge for non-smokers: ${non_smoker_avg}")
print(f"Difference: ${diff} higher for smokers\n")


```

```python
# ----------------------
# Step 5: Compare by sex
# ----------------------
def compare_by_sex(dataset):
    males = [p["charges"] for p in dataset if p["sex"] == "male"]
    females = [p["charges"] for p in dataset if p["sex"] == "female"]
    
    avg_male = round(sum(males) / len(males), 2)
    avg_female = round(sum(females) / len(females), 2)
    
    return avg_male, avg_female

male_avg, female_avg = compare_by_sex(insurance_data)
print("=== Charges by Sex ===")
print(f"Average charge for males: ${male_avg}")
print(f"Average charge for females: ${female_avg}\n")


```

```python
# ----------------------
# Step 6: Compare by region
# ----------------------
def charges_by_region(dataset):
    regions = {}
    for person in dataset:
        region = person["region"]
        if region not in regions:
            regions[region] = []
        regions[region].append(person["charges"])
    
    avg_by_region = {region: round(sum(charges)/len(charges), 2) for region, charges in regions.items()}
    return avg_by_region

region_averages = charges_by_region(insurance_data)
print("=== Average Charges by Region ===")
for region, cost in region_averages.items():
    print(f"{region}: ${cost}")
print()


```

```python
# ----------------------
# Step 7: Effect of number of children
# ----------------------
def charges_by_children(dataset):
    child_counts = {}
    for person in dataset:
        num = person["children"]
        if num not in child_counts:
            child_counts[num] = []
        child_counts[num].append(person["charges"])
    
    avg_by_children = {num: round(sum(charges)/len(charges), 2) for num, charges in child_counts.items()}
    return avg_by_children

children_averages = charges_by_children(insurance_data)
print("=== Average Charges by Number of Children ===")
for num, cost in sorted(children_averages.items()):
    print(f"{num} child(ren): ${cost}")
```
